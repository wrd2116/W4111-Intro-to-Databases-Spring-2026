### Simple Match

```python
# I want to find all the people with vocation = "Engineer"
#
#
query = {"vocation": "ENGINEER"}
result = person_coll.find(filter=query)
list(result)
```

```python
# If I only want a subset of the properties, I can also use projection.
#
query = {"vocation": "ENGINEER"}
projection = {
    "firstname" : 1,
    "lastname": 1,
    "vocation": 1,
    "_id": 0
}
result = person_coll.find(filter=query, projection=projection)
list(result)
```

### Match

I use "$match" operation is I want to "find" documents in a stage in a pipeline.

```python
pipeline = []
pipeline.append({"$match": {"vocation": "ENGINEER"}})
```

```python
pipeline
```

```python
result = person_coll.aggregate(pipeline)
```

```python
list(result)
```

```python
#
# I can also project within a pipeline
#
pipeline.append({
    "$project": {
        "firstname": 1,
        "lastname": 1,
        "vocation" : 1,
        "_id": 0
    }
})
pipeline
```

```python
result = person_coll.aggregate(pipeline)
list(result)
```
