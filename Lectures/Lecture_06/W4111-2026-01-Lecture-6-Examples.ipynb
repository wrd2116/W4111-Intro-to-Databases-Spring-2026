```python
# This is the SQLAlchemy connection library approach to connections/sessions.
#
engine = create_engine(db_url)
```

```python
# SQL query
sql = text("""
    SELECT ID, name, dept_name, tot_cred
    FROM db_book.student
    WHERE dept_name = :dept
    ORDER BY name
""")

# Execute query
with engine.connect() as conn:
    result = conn.execute(sql, {"dept": "Comp. Sci."})

    # Print results
    # You can see that SQLAlchemy surfaces the cursor concept as an interator.
    for row in result:
        print(dict(row._mapping))
```
